### Identificação e Tratamento dos Outliers
Outliers são valores que se diferenciam significativamente da maioria dos dados de um dataset, podendo influenciar de maneira desproporcional as análises estatísticas e resultados.

Inicialmente, fazemos um overview da estatística descritiva das variáveis:

A análise estatística descritiva revelou a presença de **valores fora do senso comum** em algumas variáveis. Por isso, adotaremos critérios de corte para remover registros com possíveis **outliers extremos**, com base em regras coerentes com o funcionamento da plataforma Airbnb.

###  'minimum_nights'
Essa variável possui valores que chegam até **1.124 noites**, o que representa mais de **3 anos** de estadia mínima. Como o foco deste estudo é em **locações de curta temporada**, adotaremos um limite máximo de **90 dias** (3 meses). Estadas superiores a esse valor são atípicas para esse tipo de acomodação e podem indicar uso da plataforma com outro propósito (como aluguel residencial tradicional).

Para tratar e estabelecer limites aos outliers, utilizaremos a regra do Intervalo Interquartil (IQR).

### Tratamento de Outliers com Z-Score

A distribuição da variável 'price' é **assimétrica à direita** (positivamente assimétrica), ou seja, há muitos preços baixos e poucos preços muito altos.

A técnica do **IQR (Interquartile Range)** define limites com base nos quartis, mas em distribuições assimétricas, pode resultar em um **limite superior muito conservador**, eliminando valores altos que ainda podem ser plausíveis (como diárias de hotéis ou apartamentos de luxo em Barcelona).

Para lidar com isso, podemos adotar a técnica do **Z-Score** no tratamento do limite superior.

O **Z-Score** (ou escore-z) é uma medida estatística que indica **quantos desvios padrão (DP)** um determinado valor está distante da **média** do conjunto de dados.

Ele é calculado pela seguinte fórmula:


\[
Z = \frac{X - \mu}{\sigma}
\]


Onde:
- \( X \) é o valor observado;
- \( \mu \) é a média dos dados;
- \( \sigma \) é o desvio padrão.

Essa abordagem permite **identificar outliers superiores extremos** (valores com Z-Score > 3), sem eliminar valores altos plausíveis.


```python
# Calculando o limite superior que tenha um Z-Score de 3:
# 3 = (X - µ)/ σ  
# X = 3*σ + µ
lim_price_zscore = 3 * df_tratado.price.std() + df_tratado.price.mean()
print(f"Utilizaremos {lim_price_zscore:.2f} como limite de corte para outliers superiores")
```

## Verificando a existência de correlação entre as variáveis numéricas

O coeficiente de correlação é uma medida estatística utilizada para quantificar a força e a direção da relação linear entre duas variáveis numéricas. Seu valor varia de -1 a 1, onde:

* 1 representa uma correlação positiva perfeita (ambas variáveis aumentam juntas);

* -1 indica uma correlação negativa perfeita (uma variável aumenta enquanto a outra diminui);

* 0 sugere ausência de correlação linear entre as variáveis analisadas.

```python
# Definindo a matriz de correlação
corr = df_tratado[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 
                 'calculated_host_listings_count', 'availability_365', 'number_of_reviews_ltm']].corr()
# Mostrando a matriz de correlação
display(corr);
```

```python
# Mapa de calor das correlações
sns.heatmap(corr, cmap='coolwarm', fmt= '.2f', square=True, linecolor='white', annot=True)
```

**Observações:**

* As variáveis relacionadas a avaliações (como number_of_reviews, reviews_per_month, number_of_reviews_ltm) tendem a se correlacionar fortemente entre si, o que é intuitivo e lógico.

* Preço tem baixa correlação com quase todas as variáveis, exceto uma correlação moderada negativa com minimum_nights (Aparentemente devido à quanto maior o número mínimo de noites exigido, menor tende a ser o preço. Talvez porque anúncios mais longos ofereçam descontos ou sejam menos atrativos).

* Algumas variáveis podem não ter relação direta entre si, o que pode indicar independência ou necessidade de outras análises multivariadas.

Na plotagem de histograma, observamos uma concentração acentuada para valores próximos à 5 e outra concentração para valores próximos à 30 na variável 'minimum_nights'. Vamos realizar uma comparação entre os preços desses dois grupos.

```python
# Verificando a média do preço da diária das acomodações que requerem uma estadia mínima de até 10 dias:
media_preco_5night = df_tratado[df_tratado.minimum_nights <=10]
media_preco_5night.price.mean()
```

```python
# Verificando a média de preço da diária das acomodações que requerem estadia mínima entre 25 e 35 dias:
media_preco_25a30night = df_tratado[(df_tratado.minimum_nights > 25) & (df_tratado.minimum_nights <=35)].price.mean()
media_preco_25a30night
```

A média de preço das acomodações que exigem estadia mínima entre 25 e 35 dias é aproximadamente 60% inferior à média das acomodações que permitem estadia mínima entre 1 e 5 dias.
